Problem Statement: The goal of this project is to predict the individual medical costs billed by health insurance based on various patient attributes. This is a Supervised Learning task.

Problem Type: Regression (as we are predicting a continuous numerical value).

Target Variable: charges (The specific medical insurance cost).

Features: Age, Sex, BMI, Number of children, Smoking status, and Region.

Model Goal: To learn the relationship between patient demographics/habits and their insurance charges to provide accurate cost estimates for new patients.
